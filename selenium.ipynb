```python
from bs4 import BeautifulSoup
from selenium import webdriver

```

```python
driver = webdriver.Chrome()
#create a new instance of the chrome browser


url="https://www.gramedia.com/categories/buku"
#url website
driver.get(url)
#mengarakhkan browser ke url

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
#membaca code html
print(soup)

#extract judul buku
#mencari elemen
soup.find() #akan menumukan 1 elemen saja, tapi jika elemennya ada banyak dia hanya mengembalikan 1 saja juga
titles = soup.find_all("div", {"class": "list-title"}) #akan menemukan lebih dari 1 elemen

print(titles)
```
