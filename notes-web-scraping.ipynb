```python
# import selenium dan bs4
from bs4 import BeautifulSoup
from selenium import webdriver
import time

# Create a new instance of the Chrome browser
# cara 1
# driver = webdriver.Chrome("/Users/yudanto/Downloads/chromedriver-mac-arm64")

# cara 2
driver = webdriver.Chrome()

# url website
url="https://www.gramedia.com/categories/buku"

# mengarahkan browser ke url
driver.get(url)

# delay render tampilan selama 3 detik
time.sleep(5)

# membaca code html
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

print(soup)
```

```python
# mencari elemen
# soup.find() # akan menemukan 1 elemen saja, tapi jika elemen nya ada banyak dia hanya mengembalikan 1 saja juga

# workflow 1
titles = soup.find_all("div", {"class":"list-title"}) # akan menemukan lebih dari 1 elemen
prices = soup.find_all("p", {"class":"formats-price"}) # akan menemukan lebih dari 1 elemen

# extract judul buku
for element in titles:
    print(element.get_text())

# extract harga buku
for element in prices:
    print(element.get_text())
```

```python
# workflow 2
listElement = soup.find_all("div", {"class":"entities-list"})

for item in listElement:
    # extract elemen title di setiap kotak
    title = item.find("div", {"class":"list-title"})

    # extract elemen price di setiap kotak
    price = item.find("p", {"class":"formats-price"})
    print(title.get_text() + " - " + price.get_text())
```

```python
# scraping multiple page

# import selenium dan bs4
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd

driver = webdriver.Chrome()

listTitle = []
listPrice = []

for page in range(1, 2):
    # url website
    url = f"https://www.gramedia.com/categories/buku?page={page}"

    # mengarahkan browser ke url
    driver.get(url)

    # delay 5 detik
    time.sleep(5)

    # membaca code html
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    # baca setiap elemen di setiap halaman
    listElement = soup.find_all("div", {"class": "entities-list"})

    for item in listElement:
        # extract elemen title di setiap kotak
        title = item.find("div", {"class": "list-title"})
        listTitle.append(title.get_text())  # menyimpan data title ke listTitle

        # extract elemen price di setiap kotak
        price = item.find("p", {"class": "formats-price"})
        listPrice.append(price.get_text())  # menyimpan data price ke listPrice

# bikin dataframe kosong
df = pd.DataFrame()

# membuat kolom dan isinya
df["Title"] = listTitle
df["Price"] = listPrice

# export data ke csv
df.to_csv('dataset.csv', index=False)
```
